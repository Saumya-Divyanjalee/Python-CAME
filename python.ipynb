```python
# Import NumPy library for numerical operations and array handling
import numpy as np

# Import matplotlib's pyplot module for displaying images and plots
import matplotlib.pyplot as plt

# Create a 5x5 NumPy array representing a grayscale image
# Each value ranges from 0 (black) to 255 (white)
img = np.array([
    [10, 247, 197, 55, 23],
    [123, 34, 255, 0, 170],
    [78, 90, 45, 200, 100],
    [255, 255, 100, 50, 90],
    [40, 60, 72, 84, 0]
])

# Print the original image matrix to the console
print("Original image:\n", img)

# Display the image using grayscale color mapping
plt.imshow(img, cmap='gray')

# Set the title of the displayed image
plt.title('Original Grayscale Image')

# Hide x-axis and y-axis values
plt.axis('off')
# Show the image window
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img_A = np.random.randint(0, 256, (10, 10, 3))
img_B = np.random.randint(0, 256, (10, 10, 3))
img_C = np.random.randint(0, 256, (10, 10, 3))

print("Image A shape:", img_A.shape)
print("Image B shape:",img_B.shape)
print("Image C shape:",img_C.shape)

A_flat = img_A.flatten()
B_flat = img_B.flatten()
C_flat = img_C.flatten()

dot_A_B = np.dot(A_flat, B_flat)
dot_A_C = np.dot(A_flat, C_flat)

if dot_A_C > dot_A_C:
  print("Image B is more similar to Image A")
else:
  print("Image C is more similar to Image B")


print("\nDot Product A  B:", dot_A_B)
print("Dot Product A  C:", dot_A_C)
```

```python
import numpy as np

movie_A = np.array([2.5, 0.5])
movie_B = np.array([10, 5])
user = np.array([5, 1])

dot_movie_A_user = np.dot(movie_A, user)
dot_movie_B_user = np.dot(movie_B, user)

movie_A_norm = np.linalg.norm(movie_A)
movie_B_norm = np.linalg.norm(movie_B)
user_norm = np.linalg.norm(user)

cosine_A = dot_movie_A_user / (movie_A_norm * user_norm)
cosine_B = dot_movie_B_user / (movie_B_norm * user_norm)

print("Dot Product (Movie A  User):", dot_movie_A_user)
print("Dot Product (Movie B  User):", dot_movie_B_user)

print("\nCosine Similarity (Movie A, User):", cosine_A)
print("Cosine Similarity (Movie B, User):", cosine_B)

if cosine_A > cosine_B:
    print("\ncloser to Movie A")
else:
    print("\ncloser to Movie B")
```
